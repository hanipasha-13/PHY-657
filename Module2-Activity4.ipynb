## 1. Likelihood Function

The probability density for a single data point is:

$$
p(x_n \mid \mu, \sigma^2)
=
\frac{1}{\sqrt{2\pi\sigma^2}}
\exp\left(
-\frac{(x_n-\mu)^2}{2\sigma^2}
\right)
$$

Since the data are independent, the likelihood is the product:

$$
L(\mu,\sigma^2)
=
\prod_{n=1}^{N}
\frac{1}{\sqrt{2\pi\sigma^2}}
\exp\left(
-\frac{(x_n-\mu)^2}{2\sigma^2}
\right)
$$

Simplifying:

$$
L(\mu,\sigma^2)
=
(2\pi\sigma^2)^{-N/2}
\exp\left(
-\frac{1}{2\sigma^2}
\sum_{n=1}^{N}(x_n-\mu)^2
\right)
$$



## 2. Log-Likelihood

Taking the logarithm:

$$
\ell(\mu,\sigma^2)
=
\log L
=
-\frac{N}{2}\log(2\pi)
-\frac{N}{2}\log\sigma^2
-\frac{1}{2\sigma^2}
\sum_{n=1}^{N}(x_n-\mu)^2
$$



## 3. Maximum Likelihood Estimator for $\mu$

Differentiate the log-likelihood with respect to $\mu$:

$$
\frac{\partial \ell}{\partial \mu}
=
\frac{1}{\sigma^2}
\sum_{n=1}^{N}(x_n-\mu)
$$

Set equal to zero:

$$
\sum_{n=1}^{N}(x_n-\mu)=0
$$

$$
\sum x_n - N\mu = 0
$$

Thus,

$$
\boxed{
\hat{\mu}_{\text{MLE}}
=
\frac{1}{N}\sum_{n=1}^{N} x_n
}
$$

The MLE for the mean is the **sample mean**.


## 4. Maximum Likelihood Estimator for $\sigma^2$

Differentiate the log-likelihood with respect to $\sigma^2$:

$$
\frac{\partial \ell}{\partial \sigma^2}
=
-\frac{N}{2\sigma^2}
+
\frac{1}{2\sigma^4}
\sum_{n=1}^{N}(x_n-\mu)^2
$$

Set equal to zero:

$$
- N\sigma^2 + \sum_{n=1}^{N}(x_n-\mu)^2 = 0
$$

Therefore,

$$
\boxed{
\hat{\sigma}^2_{\text{MLE}}
=
\frac{1}{N}
\sum_{n=1}^{N}
(x_n-\mu)^2
}
$$

Substituting $\mu = \hat{\mu}$:

$$
\boxed{
\hat{\sigma}^2_{\text{MLE}}
=
\frac{1}{N}
\sum_{n=1}^{N}
(x_n-\bar{x})^2
}
$$



## 5. Comparison with the Unbiased Sample Variance

The unbiased estimator of the variance is:

$$
s^2
=
\frac{1}{N-1}
\sum_{n=1}^{N}
(x_n-\bar{x})^2
$$

### Difference

- **MLE variance:** uses $\frac{1}{N}$
- **Unbiased variance:** uses $\frac{1}{N-1}$

### Why the Difference?

When estimating $\sigma^2$, we already used the data to estimate $\mu$.
This costs one **degree of freedom**.

Therefore, only $N-1$ deviations are independent.

So the MLE slightly underestimates the true variance.



## When Does This Matter?

- For **large $N$**:  N-1 and N are the same, hence the difference becomes negligible.

- For **small sample sizes**, the correction is important as N-1 and N are clearly different.
