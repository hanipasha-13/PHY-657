```python
import numpy as np, matplotlib.pyplot as plt

np.random.seed(0); M=20000
for N in [1,2,10]:
    xbar = np.random.rand(M,N).mean(1)
    h,e = np.histogram(xbar, bins=50, density=True)
    c = 0.5*(e[1:]+e[:-1]); m = h>0
    a,b,c0 = np.polyfit(c[m], np.log(h[m]), 2)
    mu, sig = -b/(2*a), np.sqrt(-1/(2*a))

    plt.figure()
    plt.hist(xbar, bins=50, density=True, alpha=.4, label=rf"hist N={N}")
    xx = np.linspace(0,1,400)
    pdf = 1/(sig*np.sqrt(2*np.pi))*np.exp(-(xx-mu)**2/(2*sig**2))
    plt.plot(xx, pdf, 'k', lw=2, label=rf"fit: $\mu={mu:.4f},\ \sigma={sig:.4f}$")
    plt.legend(); plt.xlim(0,1); plt.tight_layout()

plt.show()
```

#### Examples where a Gaussian PDF is appropriate

1. Distribution of an average of many independent measurements:
If you take repeated independent readings of something bounded or non-Gaussian, their mean is approximately Gaussian for large N.

2. Aggregated noise from many small sources:
If a measurement error is the sum of many small independent effects (electronics noise + quantization + thermal jitter + tiny environmental fluctuations), the net error is often well-modeled as Gaussian.
