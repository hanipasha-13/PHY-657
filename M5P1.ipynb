# Neural Network Forward & Backward Pass (Worked Solution)

We are given:

- Input:  
  $
  x = \begin{pmatrix} 1.0 \\ 0.5 \end{pmatrix}, \quad y = 1
  $

- Hidden layer:
$
W^{(1)} =
\begin{pmatrix}
0.5 & -0.3 \\
0.2 & 0.8
\end{pmatrix}, \quad
b^{(1)} =
\begin{pmatrix}
0.1 \\
-0.1
\end{pmatrix}
$

- Output layer:
$
w^{(2)} =
\begin{pmatrix}
0.6 \\
-0.4
\end{pmatrix}, \quad
b^{(2)} = 0.2
$

- Activation: sigmoid  
$
\sigma(z) = \frac{1}{1 + e^{-z}}
$

---

# (a) Forward Pass

## Step 1: Hidden layer pre-activation
$
z^{(1)} = W^{(1)} x + b^{(1)} = \begin{pmatrix}
0.5 & -0.3 \\
0.2 & 0.8
\end{pmatrix} \begin{pmatrix} 1.0 \\ 0.5 \end{pmatrix} + \begin{pmatrix}
0.1 \\
-0.1
\end{pmatrix}
$

$
z^{(1)} =
\begin{pmatrix}
0.45 \\
0.5
\end{pmatrix}
$

---

## Step 2: Hidden activations
$
h_i = \sigma(z^{(1)}_i)
$

$
h_1 = \sigma(0.45) \approx 0.610
$

$
\\
h_2 = \sigma(0.5) \approx 0.622
$

$
h =
\begin{pmatrix}
0.610 \\
0.622
\end{pmatrix}
$

---

## Step 3: Output pre-activation
$
z^{(2)} = w^{(2)} \cdot h + b^{(2)}
$

$
z^{(2)} = 0.6(0.610) + (-0.4)(0.622) + 0.2
$

$
= 0.366 - 0.249 + 0.2 = 0.317
$

---

## Step 4: Output
$
\hat{y} = \sigma(z^{(2)}) = \sigma(0.317) \approx 0.579
$

---

# (b) Loss (Binary Cross-Entropy)

$
L = -\left[ y \log(\hat{y}) + (1-y)\log(1-\hat{y}) \right]
$

Since \( y = 1 \):

$
L = -\log(0.579) \approx 0.546
$

---

# (c) Backward Pass

## Output layer error
$
\delta^{(2)} = \hat{y} - y = 0.579 - 1 = -0.421
$

---

## Gradients for output layer

$
\frac{\partial L}{\partial w^{(2)}} = \delta^{(2)} \cdot h
$

$
=
-0.421
\begin{pmatrix}
0.610 \\
0.622
\end{pmatrix}
=
\begin{pmatrix}
-0.257 \\
-0.262
\end{pmatrix}
$

$
\frac{\partial L}{\partial b^{(2)}} = \delta^{(2)} = -0.421
$

---

## Hidden layer error

$
\delta^{(1)} = (w^{(2)} \delta^{(2)}) \odot \sigma'(z^{(1)})
$

First:
$
w^{(2)} \delta^{(2)} =
\begin{pmatrix}
0.6 \\
-0.4
\end{pmatrix}
(-0.421)
=
\begin{pmatrix}
-0.253 \\
0.168
\end{pmatrix}
$

Sigmoid derivative:
$
\sigma'(z) = \sigma(z)(1 - \sigma(z))
$

$
\sigma'(0.45) \approx 0.610(1 - 0.610) = 0.238
$
$
\sigma'(0.5) \approx 0.622(1 - 0.622) = 0.235
$

Now:
$
\delta^{(1)} =
\begin{pmatrix}
-0.253 \cdot 0.238 \\
0.168 \cdot 0.235
\end{pmatrix}
=
\begin{pmatrix}
-0.060 \\
0.039
\end{pmatrix}
$

---

## Gradients for first layer

$
\frac{\partial L}{\partial W^{(1)}} = \delta^{(1)} x^T
$

$
=
\begin{pmatrix}
-0.060 \\
0.039
\end{pmatrix}
\begin{pmatrix}
1.0 & 0.5
\end{pmatrix}
=
\begin{pmatrix}
-0.060 & -0.030 \\
0.039 & 0.0195
\end{pmatrix}
$

$
\frac{\partial L}{\partial b^{(1)}} =
\begin{pmatrix}
-0.060 \\
0.039
\end{pmatrix}
$

---

# (d) Parameter Update

Learning rate: $ \eta = 0.1 $

## Update weights

$
w^{(2)}_{\text{new}} = w^{(2)} - \eta \frac{\partial L}{\partial w^{(2)}}
$

$
=
\begin{pmatrix}
0.6 \\
-0.4
\end{pmatrix}
-
0.1
\begin{pmatrix}
-0.257 \\
-0.262
\end{pmatrix}
=
\begin{pmatrix}
0.626 \\
-0.374
\end{pmatrix}
$

---

## Update bias

$
b^{(2)}_{\text{new}} = 0.2 - 0.1(-0.421) = 0.242
$

---

# (e) Interpretation

It is called **backpropagation** because:

- The **error signal (δ)** is propagated backward through the network  
- Starting from the **output layer → hidden layer → earlier layers**

What is propagated:
- The **gradient of the loss function**

Direction:
- **From output to input (reverse of forward pass)**

Purpose:
- To compute gradients efficiently using the **chain rule**
